```python
df=pd.read_csv("all_season_summary.csv")
```

```python
df.head()
```

```python
df.drop(df[(df["home_team"]=="RPS") | (df["away_team"]=="RPS")].index,axis=0,inplace=True)
df.drop(df[(df["home_team"]=="GL") | (df["away_team"]=="GL")].index,axis=0,inplace=True)
df.drop(df[(df["home_team"]=="Kochi") | (df["away_team"]=="Kochi")].index,axis=0,inplace=True)
df.drop(df[(df["home_team"]=="PWI") | (df["away_team"]=="PWI")].index,axis=0,inplace=True)
df=df.apply(lambda x: x.replace("KXIP","PBKS"))
```

```python
df.drop(df[(df["home_team"]=="GT") | (df["home_team"]=="LSG")].index,axis=0,inplace=True)
df.drop(df[(df["away_team"]=="GT") | (df["away_team"]=="LSG")].index,axis=0,inplace=True)
```

```python
df.drop(df[(df["home_team"]=="GT") | (df["home_team"]=="LSG")].index,axis=0,inplace=True)
df.drop(df[(df["away_team"]=="GT") | (df["away_team"]=="LSG")].index,axis=0,inplace=True)
```

```python
df1=pd.DataFrame(df.groupby(["season"],as_index=False)["winner"].value_counts())
```

```python
def visualization(team):
    temp=df1[df1["winner"]==team]
    seasons_list=list(temp["season"].unique())
    plt.figure(dpi=200,figsize=(12,6))
    sns.scatterplot(x=temp["season"],y=temp["count"])
    p=sns.lineplot(x=temp["season"],y=temp["count"],linewidth=2,color='black')
    p.set_xticks(range(min(seasons_list), max(seasons_list)+1));
    plt.xticks(rotation=90);
    plt.xlabel('Season');
    plt.ylabel('Number Of Wins');
```
